```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from speech_features import SpeechFeatures

class params:
    win_size=0.025
    win_step=0.01
    nfft=512
    nfilt=40
    ndct=36
    pre_emp=0.97
    
p = params()
```

```python
S = SpeechFeatures('mngu0_s1_0001.wav',p.win_size,p.win_step,p.nfft,p.nfilt,p.ndct,p.pre_emp)
```

```python
S.plot(S.sig)
sig_len = len(S.sig)
tot_time = len(S.sig)/S.srate
print('Signal length=', sig_len)
print('Total time=', tot_time)
```
