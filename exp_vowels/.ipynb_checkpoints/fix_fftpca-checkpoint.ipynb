## 즉, 일부분의 시그널만으로 fft를 계산하는 내 코드에 문제가 있다

---

## 여기서 일부분에 대해 fft로 계산을 해보자

```python
nframes = int((len(sig_part)/S.srate-S.win_size)//S.win_step + 1)
```

```python
mag = np.array([]).reshape(0,hp.num_freq//2+1)
for i in range(nframes):
    chunk = sig_part[i*S.frame_step:i*S.frame_step+S.frame_len].copy()
    chunk *= np.hamming(len(chunk))
    _mag = np.absolute(np.fft.rfft(chunk, hp.num_freq))
    mag = np.vstack([mag, _mag])
S.implot(mag**2)    
```
