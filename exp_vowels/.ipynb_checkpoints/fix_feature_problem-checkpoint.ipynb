```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from speech_features import SpeechFeatures
from python_speech_features import mfcc as pysf_mfcc

class hparams:
     # Params
     def __init__(self):
         self.num_mels = 40
         self.num_freq = 1025  # =nfft
         self.sample_rate = 16000
         self.frame_width = 0.020
         self.frame_shift = 0.010
         self.preemphasis = 0.97
         self.n_mfcc = 36 # =ndct
         self.nfilt = 40
        
p = hparams()        
```

## 내가 만든 speech_features.py를 재 검증하자

1. 파일 하나로 평가하기 (mngu0_s1_0001.wav) -> frame예상 개수와 실제 개수 계산하기

```python
S = SpeechFeatures('mngu0_s1_0001.wav',p.frame_width,p.frame_shift,
                  p.num_freq,p.nfilt,p.n_mfcc,pre_emp=p.preemphasis)
```

```python
S.plot(S.sig)
sig_len = len(S.sig)
tot_time = len(S.sig)/S.srate
print('Signal length=', sig_len)
print('Total time=', tot_time)
```

```python
print('frame size=', p.frame_width)
print('frame shift=', p.frame_shift)
```

```python
print('Expected frames=', (tot_time - p.frame_width)//p.frame_shift + 1)
```

```python
magspec, powspec, logspec = S.get_fft()
mymfcc = S.get_mfcc(powspec)
fftdct = S.apply_dct(logspec)
```

```python
print('magspec=', magspec.shape)
print('powspec=', powspec.shape)
print('logspec=', logspec.shape)
print('mymfcc=', mymfcc.shape)
print('fftdct=', fftdct.shape)
```

```python
S.implot(fftdct)
```

```python
mfcc_pysf = S.python_speech_feature_mfcc()
print('python_speech_feature mfcc=', mfcc_pysf.shape)
# 프레임 사이즈가 하나 더 큰 이유는, 첫 시작 프레임의 센터 타임이 0초이기 때문. 
# 즉, 가장 첫 윈도우의 절반이 0으로 걸쳐있음.
```

```python
mfcc_librosa = S.librosa_mfcc()
print('librosa mfcc=', mfcc_librosa.shape)
# librosa는 2프레임이 더 많네? 왜일까.
```

```python
S.implot(mymfcc)
```

```python
S.implot(mfcc_pysf)
```

```python
S.implot(mfcc_librosa)
```
