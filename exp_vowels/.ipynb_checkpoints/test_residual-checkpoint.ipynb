# Compute residual FFT after removing 3 PCs

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
from sklearn.decomposition import PCA
from speech_features import SpeechFeatures
from make_spkr_dict_fft import hparams
```

## Load data

```python
fft_raw = np.load('fft_raw.npy')
fft_pca = np.load('fft_pca.npy')
```

```python
def imshow(data):
    '''Plot spectrogram'''
    fig, ax = plt.subplots(facecolor='white')
    im = ax.imshow(data.T, aspect='auto', origin='lower')
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax)
```

```python
imshow(fft_raw)
```

```python
pca = PCA(n_components=3)
reduced = pca.fit_transform(fft_raw)
Xinv = pca.inverse_transform(reduced)
```

```python
imshow(Xinv)
```

```python
imshow(fft_raw - Xinv)
```
